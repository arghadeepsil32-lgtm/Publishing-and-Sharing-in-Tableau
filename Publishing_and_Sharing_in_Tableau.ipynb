1. **What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/Cloud in terms of accessibility and data security?**

Answer ▶

a. Data Security & Privacy

- Visibility - In case of tableau public, it is Everyone. Anything you publish is discoverable by the public and in case of tableau server, it is Private. Only authorized users within your organization can see it.

- Data Download - In case of tableau public, By default anyone can download your workbook and the underlying data  and in case of tableau server, it's Restricted. You control who can view, interact with, or download data.

- Row-Level Security - In case of tableau public, it is Not available. Everyone sees the same data and in case of tableau server, it's Available. Different users can see different data within the same dashboard.

b. Data Connectivity

- Tableau Public: Generally requires static files (Excel, CSV, Google Sheets). It does not support "Live" connections to private SQL databases or data warehouses.

- Tableau Server: Supports Live Connections and Scheduled Extracts from SQL Server, Snowflake, BigQuery, and many more. This ensures your dashboard updates automatically as the underlying data changes.

2. **When publishing a Tableau dashboard, what happens to a live connection compared to an extract?How does this affect data refresh and performance?**

Answer ▶

###How They Behave Upon Publishing

a.  Live Connection

- The Link: Tableau Server/Cloud acts as a "passthrough." Every time a user interacts with a filter or opens the dashboard, Tableau sends a fresh query directly to your database (SQL Server, Snowflake, etc.).

- Dependency: If your database goes offline or the network is slow, your dashboard will fail to load or feel sluggish.

b. Extract

- The Snapshot: When you publish, Tableau "bundles" a copy of the data into a high-performance compressed file format called Hyper.

- Security: Data security is now managed by Tableau Server permissions rather than the original database's Row-Level Security (unless you've configured specific RLS within Tableau).

###How Data Refreshes Work

- Live Connection:

There is no "refresh schedule." As soon as a row is added to your SQL database, it is available in Tableau the next time the dashboard is opened or the "Refresh" button is clicked in the browser.

- Extract: Scheduled

Since an extract is a static snapshot, it becomes outdated the second the source data changes.

Refresh Schedules: You must set up a task on Tableau Server/Cloud to "Refresh" the extract (e.g., every morning at 6 AM, or every hour).

3. **In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users can view, interact with, or edit a published dashboard?**

Answer ▶

Tableau administrators use Permissions, which are governed by the interplay of two specific features: Site Roles and Content Permissions.

###1. Site Roles

The Site Role is assigned to every user when they are added to the environment. it sets the maximum level of access a user can ever have on that site.

- Creator: Can create, publish, and edit everything.

- Explorer: Can interact with and "Web Edit" existing dashboards, but cannot create new data sources from scratch.

- Viewer: Can only see and interact with filters. They can never edit or save changes, even if a specific folder's settings say they can.

###2. Content Permissions

Once a user has a Site Role, administrators use Permission Rules to grant specific "capabilities" on a piece of content (like a Workbook) or a container (like a Project).

Common capabilities include:

- View: Can the user even see the dashboard in their list?

- Filter: Can they change the dropdowns and parameters?

- Web Edit: Can they click the "Edit" button to change the layout in the browser?

- Download Summary Data: Can they export the numbers behind the charts?

4. **If you want users to view a dashboard but not download the underlying data, which permission setting would you adjust in Tableau Server?**

Answer ▶

To prevent users from accessing the raw numbers behind your visualizations while still allowing them to see the dashboard, I would specifically adjust the Download Summary Data and Download Full Data capabilities.

###The Specific Settings to Adjust:

Download Summary Data: Setting this to Deny prevents users from seeing the aggregated data used to create the marks on your view (e.g., the total sales number for a specific bar in a chart).

Download Full Data: Setting this to Deny (which is often the default for most roles) ensures users cannot see the row-level underlying records (e.g., the individual transaction lines that make up that total sales number).

5. **Name two methods by which a Tableau dashboard can be embedded into a web page or website.**

Answer ▶

a. Using the Embed Code

This is the simplest, "no-code" method. When you view a dashboard on Tableau Public, Server, or Cloud, you can click the Share button and copy the provided Embed Code.

- How it works: You paste this snippet directly into your website's HTML.

- Modern Standard: As of 2022, Tableau uses the Embedding API v3 under the hood for this. It utilizes a custom HTML tag called <tableau-viz>.

- Best for: Quick integrations where you just need the dashboard to appear as it does in Tableau.

b.  Using the Embedding API (JavaScript)

This is the "developer" method, which gives you programmatic control over the dashboard.

- How it works: You link to the Tableau Embedding JavaScript library and write custom script to initialize the visualization.

- Capabilities: Unlike the basic embed code, this allows your website to "talk" to the dashboard. For example, you could build a button on your own website that triggers a filter inside the Tableau viz or exports the data to a PDF.

- Best for: Seamlessly integrating Tableau into a custom application or creating complex interactions between the web page and the data.

6. **List two best practices you should follow when preparing a dashboard for public sharing to ensure data security and privacy.**

Answer ▶

1. Data Anonymization and Aggregation

Before you even open Tableau, you should ensure the underlying dataset contains no PII (Personally Identifiable Information).

- Anonymize: Replace names, specific addresses, or ID numbers with generic labels (e.g., "Customer 1," "Customer 2").

- Aggregate: Instead of sharing individual transaction-level data, aggregate the data at a higher level. For example, instead of showing "Total spend by [User Name] on [Specific Date]," show "Total Monthly Revenue by [Region]." This prevents anyone from "reverse-engineering" the data to identify specific individuals.

2. Disable "Download Workbook" and "View Data" Permissions

Even if your dashboard looks "safe," the underlying data can still be exposed if a user downloads the source file.

- Hide Worksheets: In Tableau Desktop, hide all worksheets before publishing so users only see the final dashboard.

- Turn off Data Downloads: When publishing to Tableau Public, go to your profile settings and uncheck the option to "Allow others to download or inspect this workbook and its data." This removes the "Download" button from the toolbar, preventing users from seeing the full dataset or the logic behind your calculations.

7. **After publishing a dashboard, users report a “Data not found” error.What is the most common cause related to data source setting?**

Answer ▶

1. The Root Cause: e

- The Error: If you choose "Prompt User," every person who opens the dashboard must have their own login for the database (not just Tableau). If they don't have a SQL/Database account, they get a "Data not found" or "Connection failed" error.

- The Fix: Change the setting to "Embedded Password" or "Service Account" during the publishing process. This allows Tableau Server to use a single, saved set of credentials to fetch data on behalf of any authorized Tableau viewer.

8. **When embedding a Tableau dashboard into a secure internal portal, what two additional security measures should be implemented to ensure only authorized users can access the embedded content**

Answer ▶

1. Enable Single Sign-On (SSO) with Trusted Authentication

Even if your internal portal is secure, the Tableau dashboard itself requires its own "handshake." To prevent users from having to log in twice (once to the portal and once to Tableau), you should implement SSO.

- How it works: You configure your portal and Tableau Server to use the same Identity Provider (IdP), such as SAML, OpenID Connect, or Active Directory/Kerberos.

- Trusted Tickets: Alternatively, you can use "Trusted Authentication," where your web server "vouchers" for the user. The web server sends a POST request to Tableau, receives a unique "ticket," and appends that ticket to the dashboard URL.

- Security Value: This ensures that if a user isn't logged into your corporate network, the embedded iframe will simply show a login screen or an "Access Denied" error, even if they have the direct URL.

2. 2. Implement Row-Level Security (RLS)


- How it works: You use the USERNAM()  or FULLNAME() functions within Tableau Calculated Fields to create a "Security Filter."
For example: [Sales Rep Name] = USERNAME().

- The Database Level: If you are using a "Live Connection" to a SQL database, you can also pass the user's credentials directly to the database so the SQL engine itself filters the rows before they even reach Tableau.

- Security Value: This prevents "horizontal" data leakage. Even if a Sales Rep manages to load the dashboard, they can only see their own accounts, not the entire company’s financial data.
